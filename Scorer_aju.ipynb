```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z=np.empty((31,49,12))
l2=np.empty((31,25,47,12))

dat=np.load('C:\\Users\\jesic\\Desktop\\Jesica\\Proyecto Terminal\\Salidas servidor\\Salidas servidor\\Ajusco\\datos_2022-01-01_00.nc.npz')
lat=dat['lat_aju']
lon=dat['lon_aju']    
```

```python
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']

for i, mes in zip(range(1,13),meses):
    if i==1 or i==3 or i==5 or i==7 or i==8 or i==10 or i==12:
        
        z=np.empty((31,49,12))
        l2=np.empty((31,25,47,12))

        for j in range(1,32):
            if i==5 and j==12:
                continue
            dat=np.load('C:\\Users\\jesic\\Desktop\\Jesica\\Proyecto Terminal\\Salidas servidor\\Salidas servidor\\Ajusco\\datos_2022-%02d-%02d_00.nc.npz'%(i,j))
            z_aju=dat['z_aju']
            l2_aju=dat['l2']
            z[j-1,:,:]=z_aju
            l2[j-1,:,:,:]=l2_aju

        z_mean=np.mean(z, axis=0)
        l2_mean=np.mean(l2, axis=0)

        plt.rcParams['axes.grid'] = True
        plt.rcParams['figure.max_open_warning']=0

        fig,ax = plt.subplots(3,3, figsize=(25,20))
        fig.suptitle('%s, coord(%f,%f), altura:%f m'%(mes,lon[4],lat[4],z[0,0,4]),fontsize=30)

        time=np.arange(0,25,3)
        cont=0
        #ax[0,1].legend(time,bbox_to_anchor=(1.04, 0.5), loc="center left", borderaxespad=0) 
        for t in range(0,25,3):
            if t==0 or t==3 or t==6:
                
                ax[0,cont].plot(l2_mean[t,:,4],z_mean[1:-1,4],color='orange')
                ax[0,cont].set_title('Parametro de Scorer, hora:%d'%t)
                ax[0,cont].set_xlim(-0.015,0.25)
                ax[0,cont].set_ylim(3100,15000)
                
            if t==9 or t==12 or t==15:
                ax[1,cont-3].plot(l2_mean[t,:,4],z_mean[1:-1,4],color='orange')
                ax[1,cont-3].set_title('Parametro de Scorer, hora:%d'%t)
                ax[1,cont-3].set_xlim(-0.015,0.25)
                ax[1,cont-3].set_ylim(3100,15000)
            if t==18 or t==21 or t==24:
                ax[2,cont-6].plot(l2_mean[t,:,4],z_mean[1:-1,4],color='orange')
                ax[2,cont-6].set_title('Parametro de Scorer, hora:%d'%t)
                ax[2,cont-6].set_xlim(-0.015,0.25)
                ax[2,cont-6].set_ylim(3100,15000)
            
            cont=cont+1

        
        plt.savefig('C:\\Users\\jesic\\Desktop\\Jesica\\Proyecto Terminal\\Salidas servidor\\Salidas servidor\\frec_aju\\P_Scorer_%s.jpg'%mes)
```
