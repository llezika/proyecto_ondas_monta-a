```python
# cargar las bibliotecas que usaremos

import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
import numpy as np
import xarray as xr
from netCDF4 import Dataset
import cartopy.crs as crs
from cartopy.feature import NaturalEarthFeature
from matplotlib.cm import get_cmap
from wrf import to_np, getvar, CoordPair, vertcross, interpline
import wrf
import metpy.calc as mpcalc
from metpy.plots import SkewT
from metpy.units import units

import warnings
warnings.filterwarnings('ignore')
```

```python
meses = ['enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre']
filename = []
for i,mes in zip(range(1,13),meses):#mes 
    for j in range(1,32):
        if i==1 or i==3 or i==5 or i==7 or i==8 or i==10 or i==12:
            if i<=9 and j<=9:
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-0%d_00.nc'%(i,mes,i,j))
            elif i<=9 and j>9:
                if i==5 and j==12:
                    continue
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-%d_00.nc'%(i,mes,i,j))
            elif i>9 and j<=9:
                filename.append('/ServerData/WRF_Pronostico2/2022/%d_%s/wrfout_d02_2022-%d-0%d_00.nc'%(i,mes,i,j))
            elif i>9 and j>9:
                filename.append('/ServerData/WRF_Pronostico2/2022/%d_%s/wrfout_d02_2022-%d-%d_00.nc'%(i,mes,i,j))
            
                
    for j0 in range(1,29):
        if i==2:
            if j0<=9:
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-0%d_00.nc'%(i,mes,i,j0))
            else:
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-%d_00.nc'%(i,mes,i,j0))
                
    for j1 in range(1,31):
        if i==4 or i==6 or i==9 or i==11:
            if i<=9 and j1<=9:
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-0%d_00.nc'%(i,mes,i,j1))
            elif i<=9 and j1>9:
                filename.append('/ServerData/WRF_Pronostico2/2022/0%d_%s/wrfout_d02_2022-0%d-%d_00.nc'%(i,mes,i,j1))
            elif i>9 and j1<=9:
                filename.append('/ServerData/WRF_Pronostico2/2022/%d_%s/wrfout_d02_2022-%d-0%d_00.nc'%(i,mes,i,j1))
            elif i>9 and j1>9:
                filename.append('/ServerData/WRF_Pronostico2/2022/%d_%s/wrfout_d02_2022-%d-%d_00.nc'%(i,mes,i,j1))
            
```

```python
lat1 = 19.0
lat2 = 19.5
lon1 = -99.8
lon2 = -98.4


def calc_curvature(U,z):
    '''Calculate second derivative of U wrt z'''
    d2Udz2 = (U[2:,...] - 2*U[1:-1,...] + U[0:-2,...]) / ((z[2:,...]-z[1:-1,...])*(z[1:-1,...]-z[0:-2,...]))
    return d2Udz2

for nombre in filename:

    ncfile=Dataset(nombre)
    # cargar DataFrame con xarray
    ds = xr.open_dataset(nombre)
    print(nombre)
    x_y_1 = wrf.ll_to_xy(ncfile, lat1, lon1).data
    x_y_2 = wrf.ll_to_xy(ncfile, lat2, lon2).data

    ilat_1 = x_y_1[1]
    ilat_2 = x_y_2[1]
    ilon_1 = x_y_1[0]
    ilon_2 = x_y_2[0]

    altm_popo=[]
    latm_popo=[]
    lonm_popo=[]
    ilatm_popo=[]
    ilonm_popo=[]
    t=20
    for i in range(ilat_1,ilat_2):#lat
        for j in range(ilon_1,ilon_2):#lon
            if ds.HGT[t,i,j]>3700:
                altm_popo.append(ds.HGT[t,i,j])
                latm_popo.append(ds.XLAT[t,i,j])
                lonm_popo.append(ds.XLONG[t,i,j])
                ilatm_popo.append(i)
                ilonm_popo.append(j)

    altm_popo = np.array(altm_popo)
    latm_popo = np.array(latm_popo)
    lonm_popo = np.array(lonm_popo)
    ilatm_popo = np.array(ilatm_popo)
    ilonm_popo = np.array(ilonm_popo)


    u_wrf=np.empty((25,49,156,273))
    v_wrf=np.empty((25,49,156,273))
    th_wrf=np.empty((25,49,156,273))
    for i in range(25):
        
        u_wrf[i,:,:,:]=getvar(ncfile,'ua', timeidx=i, units='m s-1')
        v_wrf[i,:,:,:]=getvar(ncfile,'va', timeidx=i, units='m s-1')
        th_wrf[i,:,:,:]=getvar(ncfile,'th', timeidx=i, units='K')

    n_lat_popo=len(latm_popo)
    theta = np.deg2rad(17)
    #Recordemos rotar 
    z=getvar(ncfile,'z',units='m')
    u_wrf_popo=np.empty((25,49,n_lat_popo))
    th_wrf_popo=np.empty((25,49,n_lat_popo))
    z_wrf_popo=np.empty((49,n_lat_popo))
    for t in range(25):
        for k in range(49):
            for i in range(n_lat_popo):
                #print(ilatm_popo[i],ilonm_popo[i])
                c1=np.cos(theta)*u_wrf[t,k,ilatm_popo[i],ilonm_popo[i]]+np.sin(theta)*v_wrf[t,k,ilatm_popo[i],ilonm_popo[i]]
                u_wrf_popo[t,k,i] = c1

    for i in range(n_lat_popo):
        z_wrf_popo[:,i]=z[:,ilatm_popo[i],ilonm_popo[i]]
        th_wrf_popo[:,:,i]=th_wrf[:,:,ilatm_popo[i],ilonm_popo[i]]

    u_wrf_popo = u_wrf_popo*units('m/s')
    z_wrf_popo = z_wrf_popo*units.meters
    th_wrf_popo = th_wrf_popo*units.kelvin

    #CALCULO DE LA REGLA DEL 1.6
    hrs=[]
    pts=[]
    for t in range(25):
        for i in range(n_lat_popo):
            if u_wrf_popo[t,25,i]>=u_wrf_popo[t,0,i]*1.6:
                #print("se cumple a la hora %f GMT el dato %d"%(t,i))
                hrs.append(t)
                pts.append(i)

    l2_wrf = np.empty((25,47,n_lat_popo))
    N2_wrf = np.empty((25,49,n_lat_popo))
    for i in range(0,25,1):

        N2=mpcalc.brunt_vaisala_frequency_squared(z_wrf_popo,th_wrf_popo[i,:,:])
        N2_wrf[i,:,:]=N2
        d2Udz2_wrf=calc_curvature(u_wrf_popo[i,:,:],z_wrf_popo)
        l2_wrf[i,:,:] = (N2[1:-1]/u_wrf_popo[i,1:-1,:]**2)-d2Udz2_wrf/u_wrf_popo[i,1:-1,:]

    narch=('datos_%s.npz'%nombre[-16:-3])
    np.savez(narch,lat_popo=latm_popo,lon_popo=lonm_popo,ilatm_popo=ilatm_popo,
             ilonm_popo=ilonm_popo,z_popo=z_wrf_popo.magnitude,hrs_16=hrs,pts16=pts,
             l2=l2_wrf,N2=N2_wrf,u_popo=u_wrf_popo.magnitude,)

    datos=np.load(narch)
    print(narch)
    print(datos.files)
```
